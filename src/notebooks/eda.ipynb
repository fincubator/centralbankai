```python
import pandas as pd
import numpy as np
import scipy.stats as sps
```

### Data Load

```python
data_dict = {}
```

```python
data_dict["train"] = pd.read_csv("../../data/train.csv")
data_dict["train"].head()
```

```python
data_dict["test"] = pd.read_csv("../../data/test.csv")
data_dict["test"].head()
```

```python
data_dict["train"].columns.tolist()
```

```python
mapper = {
    'Содержание': 'content',
    'Сервис': "service",
    'Приоритет': "priority",
    'Статус': "status",
    'Функциональная группа': "group",
    'Крайний срок': "dt_deadline",
    'Дата обращения': "dt_query",
    'Тип обращения на момент подачи': "type_query",
    'Тип обращения итоговый': "type_final",
    'Решение             ': "solution",
    'Тип переклассификации': "type_reclassification",
    'Дата восстановления': "dt_recovery",
    'Дата закрытия обращения': "dt_close",
    'Критичность': "criticality",
    'Влияние': "impact",
    'Система': "system",
    'Место': "place"
}
```

```python
data_dict["train"] = data_dict["train"].rename(mapper, axis=1)
```

```python
data_dict["test"] = data_dict["test"].rename(mapper, axis=1)
```

```python
data_dict["train"].head()
```

```python
data_dict["test"].head()
```

```python
data_dict["train"].to_csv("train_prep.csv", index=False)
```

```python
data_dict["test"].to_csv("test_prep.csv", index=False)
```
